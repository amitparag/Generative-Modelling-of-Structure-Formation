```python
import pandas as pd
import numpy as np

import seaborn as sns



df = pd.read_csv('s13to8.csv', delimiter = ",", low_memory = False)
print(df.shape)
baryonic_matter = df.iloc[:,0:5]
mean = baryonic_matter.mean(axis = 0)
st_deviation = baryonic_matter.std(axis = 0)
cutoffs = mean + 30*st_deviation
assert mean.shape == st_deviation.shape == cutoffs.shape

df = df[df.numgalaxies < cutoffs[0]][df.massCentralBlackHole
        < cutoffs[1]][df.stellarMass < cutoffs[2]][df.velDisp
        < cutoffs[3]][df.sfr < cutoffs[4]]
print(df.shape)


df['Virial_Mass'] = df['haloViralMass'] * pow(10, -10)
df['Stellar_Mass'] = df['stellarMass']  * pow(10, -10)
df['CentralBlackHoleMass'] = df['massCentralBlackHole'] * pow(10, -8 )
g = sns.pairplot(df, x_vars=['Virial_Mass', 'haloVirialRadius'],
                 y_vars=['Stellar_Mass', 'velDisp', 'CentralBlackHoleMass'],kind="scatter", 
                diag_kind = 'kde', markers = '.', plot_kws=dict(alpha = 0.6, s= 80,
                edgecolor="lightblue", linewidth=1, size=30))
g.fig.set_size_inches(18,12)
g.fig.set_label("Relation between baryonic properties and darm matter properties")
g.fig.set_size_inches(18,12)
g.fig.set_label("Relation between baryonic properties and darm matter properties")
g.fig.set_size_inches(10,10)
g.savefig("BmDm.png")


import matplotlib.pyplot as plt
%matplotlib inline
Numgalaxies = df.iloc[:,0] 
total_halo_mass = df.iloc[:,5] * pow(10, -10)
plt.scatter(y = total_halo_mass, x= Numgalaxies, color='lightblue',marker = '*', alpha=0.5 )
plt.xlabel("Number of galaxies")
plt.ylabel("Group Mass of a halo")
plt.title(" Number of Subhalos-Halo Mass relation")
plt.savefig('shdmh.png')
```

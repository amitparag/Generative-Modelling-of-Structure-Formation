```python
#This notebook will plot the stellar mass, velocity distribution and mass of central Black Hole 
#  as a function of halo Group Mass at z = 28

# Columns 1, 2,3 as a function of column 5


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

df = pd.read_csv('s27to22.csv', delimiter = ",", low_memory = False)
df.columns
```

```python
baryonic_matter = df.iloc[:,0:5]
mean = baryonic_matter.mean(axis = 0)
st_deviation = baryonic_matter.std(axis = 0)
cutoffs = mean + 30*st_deviation
assert mean.shape == st_deviation.shape == cutoffs.shape

df = df[df.numgalaxies < cutoffs[0]][df.massCentralBlackHole
        < cutoffs[1]][df.stellarMass < cutoffs[2]][df.velDisp
        < cutoffs[3]][df.sfr < cutoffs[4]]
```

```python
# For column 1, mass of CBH vs halo Group Mass

# Mass of CBH expressed in units of 10^6 Solar Masses, while halo group mass expressed in units of 10^10 solar masses

Stellar_Mass = df.iloc[:,4] 
dark_matter_halo_mass = df.iloc[:,6]

plt.scatter(x = Stellar_Mass, y= dark_matter_halo_mass, color='navajowhite',marker = '*' )
plt.xlabel("Dark Matter Mass")
plt.ylabel("Stellar Mass")
plt.title(" Stellar-Halo Mass relation")

#plot = plt.savefig("StellarHalo mass relation.png")
```
